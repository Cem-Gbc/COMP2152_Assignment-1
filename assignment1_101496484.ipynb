```python
"""
Author: Cem Dudu
Assignment: #1
"""

# b. Create 4 variables with data types commented to the right
gym_member = "Alex Alliton"   # str
preferred_weight_kg = 20.5    # float
highest_reps = 25             # int
membership_active = True      # bool

# c. This is a dictionary where keys are strings and values are tuples of integers
workout_stats = {
    "Alex": (30, 45, 20),
    "Jamie": (40, 30, 50),
    "Taylor": (20, 25, 15)
}

# d. Using a loop to calculate total workout minutes and add new key-value pairs
friends = ["Alex", "Jamie", "Taylor"]

for friend in friends:
    total_minutes = sum(workout_stats[friend])
    workout_stats[f"{friend}_Total"] = total_minutes

# e. This is a nested (2D) list containing workout minutes for each friend
workout_list = [list(workout_stats[friend]) for friend in friends]

# f. Slice the workout_list
print("Yoga and running minutes for all friends:")
for row in workout_list:
    print(row[:2])

print("\nWeightlifting minutes for the last two friends:")
for row in workout_list[-2:]:
    print(row[2])

# g. Check if any friend's total workout minutes are >= 120
for friend in friends:
    if workout_stats[f"{friend}_Total"] >= 120:
        print(f"\nGreat job staying active, {friend}!")

# h. Allow user to input a friend's name
friend_name = input("\nEnter a friend's name: ").capitalize()

if friend_name in friends:
    print(f"Workout minutes (Yoga, Running, Weightlifting): {workout_stats[friend_name]}")
    print(f"Total workout minutes: {workout_stats[f'{friend_name}_Total']}")
else:
    print(f"Friend {friend_name} not found in the records.")

# i. Print the friend with highest and lowest total workout minutes
totals = {friend: workout_stats[f"{friend}_Total"] for friend in friends}

highest_friend = max(totals, key=totals.get)
lowest_friend = min(totals, key=totals.get)

print(f"\nFriend with the highest total workout minutes: {highest_friend}")
print(f"Friend with the lowest total workout minutes: {lowest_friend}")
```
